```python
df = df.drop_duplicates()
```

```python
df['Fuel'].value_counts()
```

```python
def simplify_fuel(x):
    x = x.lower()

    if "diesel" in x:
        return "Diesel"

    if "petrol" in x or "cng" in x:
        return "Petrol"

    if "electric" in x or "hybrid" in x:
        return "Electric"

    return "Petrol"
```

```python
df['Fuel_Clean'] = df['Fuel'].apply(simplify_fuel)
```

```python
df['Fuel_Clean'].value_counts()
```

```python
df.drop('Fuel', axis=1, inplace=True)
```

```python
df
```

```python
df['Kilometers']=df['Kilometers'].str.replace(",", "")
```

```python
df
```

```python
def clean_price(x):
    if pd.isna(x):
        return None

    x = str(x).lower().replace("rs.", "").replace(",", "").strip()

    if "lakh" in x:
        num = float(x.replace("lakh", "").strip())
        return int(num * 100000)

    if "ore"  in x or "crore" in x:
        num = float(x.replace("crore", "").replace("cr", "").strip())
        return int(num * 10000000)


    try:
        return int((x))
    except:
        return None
```

```python
df['Price_clean'] = df['Price'].apply(clean_price)
```

```python
df.drop('Price', axis=1, inplace=True)
```

```python
df['Name'].str.split(' ')
```

```python
df['Year'] = df['Name'].astype(str).str.extract(r'(\b19\d{2}\b|\b20\d{2}\b)')
df['Year'] = df['Year'].astype(float).astype('Int64')
```
